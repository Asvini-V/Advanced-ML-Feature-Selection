# Decision Tree Feature Importance

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Load dataset
dataset1 = pd.read_csv("prep.csv", index_col=None)

# Preprocessing (convert categorical to numeric)
df2 = pd.get_dummies(dataset1, drop_first=True)

# Split features and target
X = df2.drop(columns=['classification_yes'])
y = df2['classification_yes']

# Train model
model = DecisionTreeClassifier(random_state=0)
model.fit(X, y)

# Get feature importance
importance = model.feature_importances_

# Display feature importance with names
for i, v in enumerate(importance):
    print(f'Feature: {X.columns[i]}, Score: {v:.5f}')

# Plot
plt.figure(figsize=(10,5))
plt.bar(X.columns, importance)
plt.xticks(rotation=90)
plt.title("Feature Importance - Decision Tree")
plt.show()
```

# Random Forest Feature Importance

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Load dataset
dataset1 = pd.read_csv("prep.csv", index_col=None)

# Preprocessing (convert categorical to numeric)
df2 = pd.get_dummies(dataset1, drop_first=True)

# Split features and target
X = df2.drop(columns=['classification_yes'])
y = df2['classification_yes']

# Train model
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X, y)

# Get feature importance
importance = model.feature_importances_

# Display feature importance with names
for i, v in enumerate(importance):
    print(f'Feature: {X.columns[i]}, Score: {v:.5f}')

# Plot
plt.figure(figsize=(10,5))
plt.bar(X.columns, importance)
plt.xticks(rotation=90)
plt.title("Feature Importance - Random Forest")
plt.show()
```
